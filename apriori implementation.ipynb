```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
ds = pd.read_excel("D:/data/Online Retail.xlsx")
ds.head()
```

```python
ds.columns
```

```python
ds["Country"].unique()
```

```python
ds["Description"] = ds["Description"].str.strip()
```

```python
ds["InvoiceNo"].isnull().sum()
```

```python
ds["InvoiceNo"] = ds["InvoiceNo"].astype('str')
ds["InvoiceNo"]
```

```python
ds=ds[~ds['InvoiceNo'].str.contains('C')]
```

```python
basket_France = (ds[ds['Country'] =="France"] 
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo')) 
```

```python
basket_France
```

```python
def hot_encode(x):
    if (x<=0):
        return 0
    return 1
basket_France = basket_France.applymap(hot_encode)
```

```python
freq_items = apriori(basket_France,min_support =0.05,use_colnames=True)
rules = association_rules(freq_items,metric ='lift',min_threshold=1)
rules.sort_values(['confidence','lift'],ascending=[False,False])
```
